```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("place_in_galaxy.csv")
df.head()
```

```python
p_mass = df['p_mas'].values
RA_deg = df['RA_deg'].values
DEC_deg = df['DEC_deg'].values
```

```python
def X_coordinate(a, d, R):
    return R * np.cos(a) * np.cos(d)

X = X_coordinate(RA_deg, DEC_deg, p_mass)
X_mean = np.radians(np.mean(X))
X_mean
```

```python
def Y_coordinate(a,d, R):
    return R * np.cos(d) * np.sin(a)

Y=Y_coordinate(RA_deg, DEC_deg, p_mass)
Y_mean = np.radians(np.mean(Y))
Y_mean
```

```python
def Z_coordinate(R, d):
    return R* np.sin(d)
Z=Z_coordinate(p_mass, DEC_deg)
```
