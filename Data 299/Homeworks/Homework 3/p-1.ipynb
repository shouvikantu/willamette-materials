```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('Prob1.csv')

# Perform Fourier analysis
signal = data['signal'].values
freq_spectrum = np.fft.fft(signal)
freq = np.fft.fftfreq(len(signal), d=0.1)

# Exclude aliases
max_freq = 5  # Hz
freq_spectrum[(np.abs(freq) > max_freq) | (freq < 0)] = 0

# Plot frequency spectrum
plt.plot(freq, np.abs(freq_spectrum))
plt.xlim([0,4])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

```python

indices = np.argsort(np.abs(freq_spectrum))[::-1][:4]  # Get indices of two largest components
periods = 1/freq[indices]  # Calculate periods
amplitudes = np.abs(freq_spectrum[indices])  # Get amplitudes
phases = np.angle(freq_spectrum[indices]) 
```

```python
for i in range(len(indices)):
    print(f"Component {i+1}: Period = {periods[i]:.2f} s, Amplitude = {amplitudes[i]:.2f}, Phase = {phases[i]:.2f} radians")
```
